```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

def plot_biomass_values(biomass_data):
    # Create a subplot
    fig = make_subplots(rows=1, cols=1)

    # Initial year to display
    initial_year = '2010'
    color_values = biomass_data[initial_year]

    # Create initial scatter plot
    scatter = go.Scattergeo(
        lon=biomass_data['Longitude'],
        lat=biomass_data['Latitude'],
        mode='markers',
        marker=dict(
            size=16,
            color=color_values,
            colorscale='Viridis',
            colorbar=dict(title='Biomass'),
            cmin=biomass_data.min().min(),
            cmax=biomass_data.max().max()
        ),
        text=biomass_data[initial_year],
        hoverinfo='text+lon+lat'
    )

    fig.add_trace(scatter)

    # Add slider steps for each year
    steps = []
    for year in biomass_data.columns[2:]:
        step = dict(
            method='update',
            label=year,
            args=[{'marker.color': [biomass_data[year]]}],
        )
        steps.append(step)

    sliders = [dict(
        active=0,
        steps=steps
    )]

    fig.update_geos(
        projection_type='natural earth',
        showland=True, landcolor='white',
        showocean=True, oceancolor='lightblue',
    )

    fig.update_layout(
        sliders=sliders,
        title='Biomass values (2010-2017)',
        width=1000,  # Adjust the width as needed
        height=750,  # Adjust the height as needed
    )

    fig.show()

plot_biomass_values(biomass)

# Call the function with your biomass data
# plot_biomass_values(biomass)  # Replace with your actual biomass data
```
